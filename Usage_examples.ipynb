# Example 1

In this example, SAFE is run with all the default settings:
* the network is the genetic interaction similarity network from Costanzo et al., 2016

# Example 2

In this example, SAFE is run with the default network but a custom attribute file:
* the network is the genetic interaction similarity network from Costanzo et al., 2016
* the attribute file is provided explicitly using an absolute path to a tab-delimited text file
    
**Note**: check the example attribute file for format specifications. The main criteria are:
* the first column contains the node labels (in this case, yeast ORFs)
* each column is an attribute

```python
sf = safe.SAFE()
```

```python
sf.load_network()
sf.define_neighborhoods()
```

```python
sf.load_attributes(attribute_file='/Users/abaryshnikova/Lab/Datasets/safe-data/attributes/hoepfner_movva_2014_doxorubucin.txt')
```

```python
# Get a "quick & dirty" answer by decreasing the number of network randomizations for estimating the enrichment p-values 
sf.compute_pvalues(num_permutations=500)
```

```python
# Show the figure and save it to a PDF file in the current folder
sf.plot_sample_attributes(show_colorbar=True, save_fig='./doxorubicin.pdf')
```
